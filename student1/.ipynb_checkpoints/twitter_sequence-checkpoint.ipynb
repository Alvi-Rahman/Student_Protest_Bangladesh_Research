```python
def countFrequency(my_list): 
  
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    return freq

        

freq_dict = countFrequency(all_topics)


for key, value in freq_dict.items(): 
        print ("% d : % d"%(key, value)) 
```
