```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('follower_followed.xlsx')
```

```python
data.tail()
```

```python
followed = list(data['followed'])
```

```python
def countFrequency(my_list): 
  
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    return freq

        

freq_dict = countFrequency(followed)
```

```python
len(freq_dict)
```

```python
len(followed)
```

```python
import operator
```

```python
freq_dict = sorted(freq_dict.items(), key=operator.itemgetter(1),reverse=True)
```

```python
freq_dict
```

```python
df = pd.DataFrame(freq_dict,columns=['id','count'])
```

```python
df.to_excel('freq.xlsx')
```
