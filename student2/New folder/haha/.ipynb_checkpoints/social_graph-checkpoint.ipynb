```python
user_id_lst = list(bnpjamaat['user_id'].unique())
```

```python
follower = pd.read_excel('follower_followed.xlsx',index_col=0)
follower_bj = follower.loc[user_id_lst].copy()
follower_bj.dropna(inplace=True)
follower = list(follower_bj.index)


followed = pd.read_excel('follower_followed.xlsx',index_col=1)
followed_bj = followed.loc[user_id_lst].copy()
followed_bj.dropna(inplace=True)
followed = list(followed_bj.index)


followed = list(set(followed))
follower = list(set(follower))
```

```python
bj_dict = {}
for i in range(len(followed)):
    if followed[i] in follower:
        bj_dict[followed[i]] = True

        
for i in range(len(follower)):
    if follower[i] in followed:
        bj_dict[follower[i]] = True
```

```python
len(followed)
```

```python
len(follower)
```

```python
len(bj_dict)
```

```python
all_bj = list(bj_dict.keys())
not_in_bj = [followed[i] for i in range(len(followed)) if followed[i] not in all_bj]
not_in_bj.extend([follower[i] for i in range(len(follower)) if follower[i] not in all_bj])
not_in_bj
```
